```python

# STEP 1: Install all required packages
!pip install -q langchain langchain-community faiss-cpu wikipedia openrouter

# STEP 2: Imports and Setup
import os
import wikipedia
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.docstore.document import Document
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain_community.llms import OpenRouter

# STEP 3: Set your OpenRouter API Key
os.environ["OPENROUTER_API_KEY"] = "your-openrouter-api-key"  # Replace with your key

# STEP 4: Fetch and split Wikipedia content
def get_wikipedia_docs(topic, max_sections=5):
    try:
        content = wikipedia.page(topic).content
        sections = content.split("\n\n")[:max_sections]
        return [Document(page_content=section) for section in sections if section.strip()]
    except Exception as e:
        print("Error fetching topic:", e)
        return []

# STEP 5: Create vector DB
def create_vector_store(documents):
    embeddings = HuggingFaceEmbeddings()
    db = FAISS.from_documents(documents, embeddings)
    return db.as_retriever()

# STEP 6: Create LLM-powered QA system
def create_qa_chain(retriever):
    llm = OpenRouter(
        model="mistralai/mistral-7b-instruct",  # You can also try llama3, falcon, etc.
        openrouter_api_key=os.getenv("OPENROUTER_API_KEY")
    )

    prompt = PromptTemplate(
        input_variables=["context", "question"],
        template="""
You are a helpful assistant. Use the following context from Wikipedia to answer the question.

Context:
{context}

Question:
{question}

Answer:"""
    )

    chain = RetrievalQA.from_chain_type(
        llm=llm,
        retriever=retriever,
        chain_type="stuff",
        chain_type_kwargs={"prompt": prompt}
    )
    return chain

# STEP 7: Main - Ask user for topic and question
topic = input("Enter a Wikipedia topic: ")
documents = get_wikipedia_docs(topic)

if not documents:
    print("No documents found for the topic.")
else:
    retriever = create_vector_store(documents)
    qa_chain = create_qa_chain(retriever)

    while True:
        query = input("\nAsk a question (type 'exit' to quit): ")
        if query.lower() in ['exit', 'quit']:
            break
        answer = qa_chain.run(query)
        print("Answer:", answer)
```
